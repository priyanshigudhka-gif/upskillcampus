```python
plt.figure(figsize=(12,6))
sns.barplot(x='Crop', y='Yield (Quintal/ Hectare) ', data=df, errorbar=None)
plt.xticks(rotation=45)
plt.title("Crop-wise Yield")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='State', y='Yield (Quintal/ Hectare) ', data=df, errorbar=None)
plt.xticks(rotation=90)
plt.title("State-wise Yield")
plt.show()
```
